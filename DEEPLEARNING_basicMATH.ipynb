## 일차함수
y = ax + b
그래프 상에서 직선인 그래프

```python
def linear_function(x):
    a = 0.5
    b = 2

    return a * x + b
```

```python
print(linear_function(5))
```

```python
x = np.arange(-5, 5, 0.1)  #-5~5까지 step은 0.1
y = linear_function(x)

plt.plot(x, y)  #x축의 데이터, y축의 데이터
#얘들은 보기 좋게 해주는
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Function')
```

## 이차함수

u = ax^2 + bx + c  일반적으로 두 개의 실근을 가짐

```python
def quadratic_function(x):
    a = 1
    b = -1
    c = -2

    return a*x**2 + b*x + c
```

```python
print(quadratic_function(2))
```

```python
x = np.arange(-5, 5, 0.1)
y = quadratic_function(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic Function')
plt.show()
```

## 미분
어떤 한순간의 변화량을 나타낸 것

기울기는 방향성을 가짐

미분식에서 h는 아주 작은 수를 뜻하는데, 10e-50 정도의 수를 하면 파이썬은 이를 0.0으로 인식
따라서, 딥러닝에서 아주 작은 수를 정할 때 1e-4 정도로 설정해도 무방

```python
def numerical_differential(f, x):  #f`(1)
    h = 1e-4
    return (f(x+h)-f(x-h)) / (2*h)
```

## 편미분

변수가 1개짜리인 위의 수치미분과 달리, 변수가 2개 이상일 때의 미분법을 편미분이라 함

다변수 함수에서 특정 변수를 제외한 나머지 변수는 상수로 처리하여 미분을 하는 것

각 변수에 대해 미분 표시를 σ(시그마)를 통해 나타남

ex) $f(x_0, x_1) = x_0 ^2 + x_1 ^2$

$f(x, y) = x^2 + y^2 - 25$라고 둠.

$x$로 편미분 ($f_x$): $2x$($y$는 상수 취급)

$y$로 편미분 ($f_y$): $2y$($x$는 상수 취급)

공식 적용: $\mathbf{\frac{dy}{dx} = -\frac{2x}{2y} = -\frac{x}{y}}$

```python
def my_func3(x):
    return (x[0]**2 + x[1]**2)
```

```python
from mpl_toolkits.mplot3d import Axes3D  #3D로 보여주기 위함

x = np.arange(-4.0, 4.0, 0.4)
y = np.arange(-4.0, 4.0, 0.4)
X, Y = np.meshgrid(x, y)  # 3차원 그래프를 그리면서 그물망 형태의 좌표쌍을 만듬.
#X, Y에는 2차원 평면위의 모든 점들을 나타내기 위한 좌표 행렬이 나타남
Z = X**2 + Y**2

fig = plt.figure()  #빈창 만들기. 도화지 생성
#ax = Axes3D(fig)
ax = fig.add_subplot(111, projection='3d') # 111: 1행1열의 첫번째 그래프. 3차원 축 생성
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='coolwarm')  #디자인
ax.set_title("f(x, y) = x**2 + y**2")
plt.show()
```

## 기울기

방향성을 가짐

```python
def numerical_diff(f, x):
    h = 1e-5
    grad = np.zeros_like(x)  # x와 모양은 같지만, 0으로 채워져 있는
    for i in range(x.size):  #x0과 x1을 따로따로 미분하기 위한 방법
        tmp = x[i]
        x[i] = tmp + h
        fxh1 = f(x)

        x[i] = tmp - h
        fxh2 = f(x)

        grad[i] = (fxh1 - fxh2) / (2*h)
        x[i] = tmp

    return grad
```

```python
print(numerical_diff(my_func3, np.array([3.0, 4.0])))
print(numerical_diff(my_func3, np.array([1.0, 2.0])))

#출력물은 방향을 뜻함. Vector. x0으로 6만큼, x1으로 8만큼
#2차원에서의 기울기
```

## 기울기의 의미를 그래프로 확인

기울기가 가장 낮은 장소(가운데)로부터 거리가 멀어질수록 기울기가 커짐

기울기가 커진다는 것은 영향을 많이 받는다는 의미

기울기가 작다는 것은 영향을 적게 받는다는 의미

```python
X = np.arange(-20, 20, 2)
Y = np.arange(-20, 20, 2)
U, V = np.meshgrid(X, Y)

fig, ax = plt.subplots()
q = ax.quiver(X, Y, U, V)  #X, Y는 화살표가 시작될 좌표. U,V는 화살표가 가르킬 방향과 크기
ax.quiverkey(q, X=0.4, Y=1.0, U=20, label='Quiver key', labelpos='E')
plt.grid()
```
